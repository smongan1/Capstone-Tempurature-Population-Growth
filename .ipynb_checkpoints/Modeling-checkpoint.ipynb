```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.metrics import median_absolute_error, mean_absolute_error, mean_squared_error
```

```python
f = open('.//Output//features.csv','r')
features = f.read()
features = features.split('\n')
f.close()
X_train = pd.read_csv('.//Output//X_train.csv',names = features).drop(columns = '')
X_test = pd.read_csv('.//Output//X_test.csv',names = features).drop(columns = '')
y_train = pd.read_csv('.//Output//y_train.csv',names = ['avg_5_year_pct_growth'])
y_test = pd.read_csv('.//Output//y_test.csv',names = ['avg_5_year_pct_growth'])
```

<h3> Simple Linear Model </h3>

```python
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
print(
    lr_model.score(X_train,y_train),
lr_model.score(X_test,y_test)
     )
```

<h3> Lasso Regression Model </h3>

```python
lasso_model = Lasso()
parameters = {
    'alpha' : [.05*(n) for n in range(1,100)], 
    'fit_intercept': [True,False], 
    'precompute' : [True,False], 
    'warm_start': [True,False]
        }
model = GridSearchCV(lasso_model,parameters,n_jobs = -1)
model.fit(X_train,y_train)
print('The score for the training set is:',
    round(model.score(X_train,y_train.values.ravel()),4),'\n',
      'The score for the test set is:',
round(model.score(X_test,y_test.values.ravel()),4)
     )
```

```python
lasso_model = model.best_estimator_
```

<h3> Random Forest Regression Model </h3>

```python
rf_model = RandomForestRegressor()
n_estimators = [31 + n**2 for n in range(2,20,3)]
max_depth = [2,5,10, None]
min_samples_split = [2,5,10]
min_samples_leaf = [1,2,5]
parameters = {
    'n_estimators' : n_estimators, 
    'max_depth': max_depth, 
    'min_samples_split' : min_samples_split, 
    'min_samples_leaf' : min_samples_leaf, 
    'warm_start': [True,False],
    'random_state' : [42]
        }
model = GridSearchCV(rf_model,parameters, n_jobs = -1)
model.fit(X_train,y_train.values.ravel())
print('The score for the training set is:',
    round(model.score(X_train,y_train.values.ravel()),4),'\n',
      'The score for the test set is:',
round(model.score(X_test,y_test.values.ravel()),4)
     )
```

```python
rf_model = model.best_estimator_
```

<h3> Gradient Boost Regression Model </h3>

```python
gb_model = GradientBoostingRegressor()
n_estimators = [31 + n**2 for n in range(2,20,3)]
loss = ['squared_error', 'absolute_error', 'huber', 'quantile']
criterion = ['friedman_mse', 'squared_error', 'mse', 'mae']
max_depth = [2,5,10, None]
min_samples_split = [2,5,10]
min_samples_leaf = [1,2,5]
parameters = {
    'n_estimators' : n_estimators, 
    'loss' : loss,
    'criterion' : criterion,
    'max_depth': max_depth, 
    'min_samples_split' : min_samples_split, 
    'min_samples_leaf' : min_samples_leaf, 
    'warm_start': [True,False],
    'random_state' : [42]
        }
model = GridSearchCV(gb_model,parameters, n_jobs = -1)
model.fit(X_train,y_train.values.ravel())
print('The score for the training set is:',
    round(model.score(X_train,y_train.values.ravel()),4),'\n',
      'The score for the test set is:',
round(model.score(X_test,y_test.values.ravel()),4)
     )
```

```python
gb_model = model.best_estimator_
```

```python
models = [lasso_model, gb_model, rf_model]
model_names = ['Lasso', 'Gradient Boost', 'Random Forest']
scorers = [median_absolute_error, mean_absolute_error,mean_squared_error, None]
scorer_names = ['Median AE', 'Mean AE', 'MSE','Temperature Importance']
scoring_table = pd.DataFrame(index = scorer_names)
for i,model in enumerate(models):
    scores = []
    y_pred = model.predict(X_test)
    for scorer in scorers:
        if scorer == None:
            if i != 0:
                importances = {features[i] : x for i,x in enumerate(model.feature_importances_)}
                score = importances['AvgTemperature']
            else:
                score = np.nan
        else:
            score = scorer(y_test,y_pred)
        scores.append(score)
    scoring_table[model_names[i]] = scores
print(scoring_table)        
```

```python
importances = [[x,features[i]] for i,x in enumerate(rf_model.feature_importances_)]
importances = sorted(importances,reverse = True)[0:11]
plt.bar([x[1] for x in importances],[x[0] for x in importances])
plt.xticks(rotation = 90)
plt.show()
```

```python
importances = [[x,features[i]] for i,x in enumerate(gb_model.feature_importances_)]
importances = sorted(importances,reverse = True)[0:11]
plt.bar([x[1] for x in importances],[x[0] for x in importances])
plt.xticks(rotation = 90)
plt.show()
```

```python
models = [lasso_model, gb_model, rf_model]
model_names = ['Lasso', 'Gradient Boost', 'Random Forest']
x_plot = [1*(k - 10) for k in range(41)]
affects_of_temp_change_table = pd.DataFrame()
for i,model in enumerate(models):
    y_pred = model.predict(X_test)
    scores = []
    for k in range(41):
        X_temp = X_test.copy()
        X_temp['AvgTemperature'] = X_temp['AvgTemperature']*(1+.01*(k - 9))
        y_temp = model.predict(X_temp)
        y_change = y_temp - y_pred
        scores.append(y_change.mean())
    affects_of_temp_change_table[model_names[i]] = scores
affects_of_temp_change_table['Percent Temperature Change'] = x_plot
atc_melt = affects_of_temp_change_table.melt('Percent Temperature Change', var_name='Model', value_name='Change')
sns.scatterplot(data = atc_melt, x = 'Percent Temperature Change', y = 'Change', hue = 'Model')
plt.show()
```
